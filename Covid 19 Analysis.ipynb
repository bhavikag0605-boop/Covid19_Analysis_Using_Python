## Active Cases

```python
covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df
```

```python
statewise = pd.pivot_table(covid_df, values = ["Confirmed", "Deaths", "Cured"], 
                           index="State/UnionTerritory", aggfunc=max)
```

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
top_10_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = ['Deaths'],ascending = False).reset_index()

fig = plt.figure(figsize=(18,5))
plt.title("Top 10 states with most Deaths", size=25)

ax = sns.barplot(data = top_10_deaths.iloc[:12], y = 'Deaths', x = 'State/UnionTerritory', linewidth=2, edgecolor = 'black') #palette='red'

plt.xlabel('States')
plt.ylabel('Total Deaths')
plt.show()
```

## Remove rows where states = India

```python
vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns = {'Total Individuals Vaccinated' : 'Total'}, inplace = True)
vaccine.tail()
```

## Most vaccinated States

```python
max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending = False)[:5]
max_vac
```

```python

fig = plt.figure(figsize=(10,5))
plt.title("Top 5 Vaccinated states in India", size=15)

x = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth=2, edgecolor = 'black') #palette='red'

plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```

## Group by State and sum the Total vaccinations

```python

min_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')

# Sort in ascending order (lowest to highest) and take the first 5
min_vac = min_vac.sort_values('Total', ascending = True)[:5]

# Display the result
min_vac
```

```python

fig = plt.figure(figsize=(15,5))
plt.title("Least 5 Vaccinated states in India", size=15)

x = sns.barplot(data = min_vac.iloc[:10], y = min_vac.Total, x = min_vac.index, linewidth=2, edgecolor = 'black') #palette='red'

plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```
